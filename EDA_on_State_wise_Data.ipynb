```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas_profiling import ProfileReport
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from fbprophet import Prophet
```

```python
dataset= pd.read_csv("/content/State_Wise_Vaccination 1.1.csv")
dataset.head()
```

```python
print("The shape of the first dataset is",dataset.shape)
```

```python
dataset.info()
```

```python
dataset.describe(include = 'all')
```

```python
dataset.isnull().sum()
```

```python
df = dataset.sort_values(by = ['Dose_1'], ascending = False)
fig = px.line(df, x = 'State_or_UT', y = 'Dose_1', title = "First Dose",
             labels = {'Dose_1': 'Number of First Dose administered'})
fig.update_layout(width=1300, height=600)
fig.show()
```

```python
df = dataset.sort_values(by = ['Dose_2'], ascending = False)
fig = px.line(df, x = 'State_or_UT', y = 'Dose_2', title = "Second Dose",
             labels = {'Dose_2': 'Number of Second Dose administered'})
fig.update_layout(width=1300, height=600)
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=dataset['State_or_UT'], y=dataset['Dose_1'],
                    mode='lines+markers',
                    name='First Dose'))
fig.add_trace(go.Scatter(x=dataset['State_or_UT'], y=dataset['Dose_2'],
                    mode='lines+markers',
                    name='Second Dose'))
fig.update_layout(width=1300, height=700)
fig.show()
```

```python
total = dataset.groupby('State_or_UT')['Cumulative_Dose'].max().reset_index()
fhc = total.sort_values('Cumulative_Dose', ascending=False).head(10)
pd.set_option('display.float_format', lambda x: '%.0f'% x)
fhc
```

```python
fig = px.bar(fhc, 
             x='State_or_UT', 
             y='Cumulative_Dose',
             labels = {'State_or_UT' : 'State_or_UT', 'Cumulative_Dose' : 'Total Vaccinations'},
             title = "Top 10 State_or_UT With Biggest Vaccinations Progress"
            )
fig.show()
```

```python
flc = total.sort_values('Cumulative_Dose', ascending=False).tail(10)
flc
```

```python
fig = px.bar(flc, 
             x='State_or_UT', 
             y='Cumulative_Dose',
             labels = {'State_or_UT' : 'State_or_UT', 'Cumulative_Dose' : 'Total Vaccinations'},
             title = "Top 10 Countries With Lowest Vaccinations Progress"
            )
fig.show()
```

```python
df = dataset.sort_values('Dose_1', ascending = False)[:10]
fig = go.Figure(data=[go.Bar(x = df['State_or_UT'], y = df['Dose_1'], text = df['Dose_1'])])
fig.update_traces(marker_color='rgb(100,200,200)', marker_line_color='rgb(0,0,0)',
                  marker_line_width=2, opacity=0.5, texttemplate = '%{text:.2s}', textposition = 'outside')
title=('Top 10 states with the most number of First Dose administered')
fig.show()
```
